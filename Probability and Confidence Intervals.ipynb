```python
plt.figure(figsize=(15,4))
plt.plot(glaxo.Close)
plt.xlabel('Date')
plt.ylabel('Closing price')
plt.title('Date vs Closing price for Glaxo')
```

```python
plt.figure (figsize=(15,4))
plt.plot(beml.Close)
plt.xlabel('Date')
plt.ylabel('Closing price')
plt.title('Date vs Closing Price for BEML')
```

```python
beml['Gain'] = beml.Close.pct_change(periods= 1)
```

```python
glaxo['Gain'] = glaxo.Close.pct_change(periods = 1)
```

```python
glaxo = glaxo.dropna()
```

```python
beml = beml.dropna()
```

```python
plt.figure (figsize=(15,4))
plt.plot(beml.index, beml.Gain)
plt.xlabel('Closing Date')
plt.ylabel('Gain %')
plt.title('Gain% vs Date for BEML')
```

```python
plt.figure (figsize=(15,4))
plt.plot (glaxo.index, glaxo.Gain)
plt.xlabel ('Closing Date')
plt.ylabel ('Gain %')
plt.title ('Gain % vs Date for GLAXO')
```

```python
plt.figure(figsize=(15,4))
sns.distplot(beml.Gain, label = 'BEML')
plt.ylabel ('Density')
plt.title ('distplot for gain')
plt.legend()
```

```python
plt.figure(figsize=(15,4))
sns.distplot(glaxo.Gain, label='GLAXO')
plt.ylabel('Density')
plt.legend()
plt.title('distplot for GLAXO')
```

```python
print('Mean:', round (glaxo.Gain.mean(), 5))
print('Standard Devidation:', round (glaxo.Gain.std(),5))
```

```python
print ("Mean:", round (beml.Gain.mean(), 5))
print ('Standard Deviation:', round (beml.Gain.std(),5))
```

```python
from scipy import stats
```

```python
1-stats.norm.cdf(0.02,
              loc=glaxo.Gain.mean(),
              scale=glaxo.Gain.std()
              )
```

```python
stats.norm.cdf(-0.02,
              loc=glaxo.Gain.mean(),
              scale=glaxo.Gain.std()
              )
```
